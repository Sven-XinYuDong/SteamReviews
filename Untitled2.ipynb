```python
import pandas as pd


assert type(friendlist_to_dataframe(extracted)) == pd.core.frame.DataFrame
assert len(friendlist_to_dataframe(extracted)) == 9
assert "matt" in friendlist_to_dataframe(extracted).values
```

```python
dataframe = friendlist_to_dataframe(extracted)
```

```python

assert chart_dataframe(dataframe)["Text"][0] == "Total number of games owned"
assert len(chart_dataframe(dataframe)) == 36
assert "TotalNumGames" in chart_dataframe(dataframe)["variable"].values
assert chart_dataframe(dataframe).columns.tolist() == ['Name', 'ProfilePicture', 'GameList', 'variable', 'value', 'Text']

```
